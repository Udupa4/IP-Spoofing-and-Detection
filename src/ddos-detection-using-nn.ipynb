```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## 1. Dataset

```python
data = pd.read_csv("/kaggle/input/ddos-sdn-dataset/dataset_sdn.csv")
data.head()
```

## 2. Data Pre-Processing

```python
df1 = data.copy(deep=True)
df1 = df1.dropna()
df1.head()
```

```python
m,n = df1.shape
print(f"No. of training examples: {m}")
print(f"No. of features: {n-1}") # (n-1) because last column is label/ target
```

```python
y = df1["label"]
y.head()
```

```python
X = df1.drop(['dt','src','dst','label'], axis=1)
X.head()
```

```python
# convert categorial data into dummy variables
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, test_size=0.2, random_state=46
)
X_train.head()
```

## 3. Importing Required Classes from TensorFlow

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## 4. Normalizing the data

```python
norm_layer = tf.keras.layers.Normalization(axis=-1)
norm_layer.adapt(X_train)  # learns mean and variance implicitly
X_norm = norm_layer(X_train)
```

```python
X_norm
```

## 5. Building the NN Model

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.Input(shape=(21,)),
        Dense(10, activation="relu", name="Layer1"),
        Dense(10, activation="relu", name="Layer2"),
        Dense(1, activation="sigmoid", name="Layer3")
    ], name="my_model"
)
```

```python
model.summary()
```

```python
# Initial Weights and bias
W1, b1 = model.get_layer("Layer1").get_weights()
W2, b2 = model.get_layer("Layer2").get_weights()
W3, b3 = model.get_layer("Layer3").get_weights()
print(f"W1: {W1.shape}")
print(f"b1: {b1.shape}")
print(f"W2: {W2.shape}")
print(f"b2: {b2.shape}")
print(f"W3: {W3.shape}")
print(f"b3: {b3.shape}")
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
    loss = tf.keras.losses.BinaryCrossentropy()
)

model.fit(
    X_norm,
    y_train,
    epochs=15,
)
```

## 6. Prediction and Accuracy using Test Set

```python
# normalize the test set
X_test_norm = norm_layer(X_test)
X_test_norm
```

```python
predictions = model.predict(X_test_norm)
y_hat = (predictions >= 0.5).astype(int)
```

```python
print(y_test[0:10])
print(y_hat[0:10])
```

```python
# Accuracy 
m = tf.keras.metrics.BinaryAccuracy()
m.reset_state()
m.update_state(y_test, predictions)
accuracy = m.result().numpy()
print(f"Accuracy achieved: {accuracy:0.4f}")
```
